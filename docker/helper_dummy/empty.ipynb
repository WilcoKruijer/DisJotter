```python
import matplotlib.pyplot as plt
```

A line from $(0, 0.25)$ to $(1, 1.5)$.

```python
line = [0.25, 1.5]
```

First calculate the mean X coordinate of the given line, then plot both lines.

```python
mean_line = sum(line) / len(line)

plt.plot(line)
plt.plot([mean_line] * 2)
plt.axis([0, 1, 0, 2])
plt.title(f'line = {line}\nmean_line = {mean_line}\n')
plt.show()
```
